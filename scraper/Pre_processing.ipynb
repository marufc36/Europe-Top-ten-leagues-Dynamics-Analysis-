```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("bundesliga_match_data.csv")



df1
```

```python
df1[['home_team_score', 'away_team_score']] = df1['score'].str.split(' - ', expand=True)
df1['home_team_score'] = pd.to_numeric(df1['home_team_score'], errors='coerce')
df1['away_team_score'] = pd.to_numeric(df1['away_team_score'], errors='coerce')
df1['home_team_score'] = df1['home_team_score'].fillna(0)
df1['away_team_score'] = df1['away_team_score'].fillna(0)




df1
```

```python
summary_data = {
    'Team': [],
    'Goals_Scored': [],
    'Goals_Conceded': [],
    'Matches_Played': [],
    'Wins': [],
    'Draws': [],
    'Losses':[]
}
summary_df = pd.DataFrame(summary_data)

teams = df1['home_team'].append(df1['away_team']).unique()
for team in teams:
    team_matches_home = df1[df1['home_team'] == team]
    team_matches_away = df1[df1['away_team'] == team]
    team_matches = pd.concat([team_matches_home, team_matches_away])

    goals_scored = team_matches['home_team_score'].sum()
    goals_conceded = team_matches['away_team_score'].sum()
    matches_played = len(team_matches)
    wins = len(team_matches[team_matches['home_team_score'] > team_matches['away_team_score']])
    draws = len(team_matches[team_matches['home_team_score'] == team_matches['away_team_score']])

    losses = len(team_matches[team_matches['home_team_score'] < team_matches['away_team_score']])

    summary_df = summary_df.append({
        'Team': team,
        'Goals_Scored': goals_scored,
        'Goals_Conceded': goals_conceded,
        'Matches_Played': matches_played,
        'Wins': wins,
        'Draws':draws,
        'Losses': losses
    }, ignore_index=True)
```

```python
summary_df['Country']='Germany'
summary_df['League']='Bundesliga'
summary_df
```

```python
summary_df.to_csv('Bundesliga.csv', index=False)
```
